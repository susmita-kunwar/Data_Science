Overlaying bar plot on top of each other

```python
# Create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])

# Plot bars in stacked manner
plt.bar(x, y1, color='r', label='Round 1')
plt.bar(x, y2, bottom=y1, color='b', label='Round 2')
plt.bar(x, y3, bottom=y1 + y2, color='y', label='Round 3')
plt.bar(x, y4, bottom=y1 + y2 + y3, color='g', label='Round 4')

# Add labels and title
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend()
plt.title("Scores by Teams in First Rounds")

# Show plot
plt.show()
```

```python
# Create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])

# Plot bars in stacked manner
plt.barh(x, y1, color='r', label='Round 1')
plt.barh(x, y2, left=y1, color='b', label='Round 2')
plt.barh(x, y3, left=y1 + y2, color='y', label='Round 3')
plt.barh(x, y4, left=y1 + y2 + y3, color='g', label='Round 4')

# Add labels and title
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend()
plt.title("Scores by Teams in First Rounds")

# Show plot
plt.show()
```
