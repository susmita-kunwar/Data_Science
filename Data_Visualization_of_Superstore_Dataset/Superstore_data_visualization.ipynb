```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Superstore.csv", encoding= 'latin-1')
df.head(15)
```

## Line Chart

```python
## Convert the 'Order Date' column to datetime formate.
df["Order Date"] = pd.to_datetime(df["Order Date"], format= "%m/%d/%Y")
df['Order Date']

## Extract the month and year from 'Order Date'.
df['Month-Year'] = df['Order Date'].dt.strftime('%Y-%m') 
df["Month-Year"] 

# Group by Year-Month and calculate total sales
monthly_sales = df.groupby('Month-Year')['Sales'].sum().reset_index()
monthly_sales.head(5)

## Create a line chart of Monthly Sales Trend.
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Month-Year'], monthly_sales['Sales'], marker= 'o', linestyle= '-', color= 'b')

## Customize your plot with title, labels, grid, and markers.
plt.title("Monthly Sales Trend", fontsize=14)
plt.xlabel("Year-Month", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)

plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Bar Chart

```python
df["Region"]
```

```python
## Group the data by 'Region' and calculate total Sales.
Group_Region_calculate_total_sales = df.groupby("Region")['Sales'].sum().reset_index()
Group_Region_calculate_total_sales

## Create a bar chart of Sales by Region.
plt.figure(figsize=(10, 6))
colors = ['Pink', 'blue', 'green', 'purple']
plt.bar(Group_Region_calculate_total_sales['Region'], Group_Region_calculate_total_sales['Sales'], color= colors)

## Add different colors for bars, title, axis labels, and value labels.

plt.title("Total Sales by Region", fontsize=14)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Pie Chart

```python
## Group the data by 'Category' and calculate total Profit.
Total_Profit_by_Category = df.groupby("Category")["Profit"].sum().reset_index()
Total_Profit_by_Category

# Define colors
colors = ['blue', 'green', 'pink']

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(Total_Profit_by_Category['Profit'], labels=Total_Profit_by_Category['Category'], 
        autopct='%.2f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})

# Add title
plt.title("Profit Share by Category", fontsize=14)

# Show the plot
plt.show()
```

## Histogram

```python
## Use the 'Sales' column to plot a histogram.
df["Sales"] = df["Sales"].astype(int)

plt.hist(df['Sales'], bins= 10, edgecolor= 'white', color= 'blue')

# Add title and labels
plt.title("Distribution of Sales", fontsize=14)
plt.xlabel("Sales Amount", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Show grid
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group data by Region and Category
sales_data = df.groupby(['Region', 'Category'])['Sales'].sum().unstack()

# Plot stacked bar chart
ax = sales_data.plot(kind='bar', stacked=True, figsize=(8, 5))

# Add title and labels
plt.title("Sales by Region and Category")
plt.xlabel("Region")
plt.ylabel("Total Sales")

# Show legend (before plt.show())
plt.legend(title="Category", loc='best', bbox_to_anchor=(0.3, 0.7))
plt.show()
```
