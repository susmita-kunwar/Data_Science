```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("IPL Data.csv")
df.head(10)
```

## Handling Missing Values

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Convert Year to integer (it's currently a float)
df['Year'] = df['Year'].fillna(0).astype(int)
df.head()
```

```python
df.isnull().sum()
```

```python
# covert datatype object to numeric so that it will be easier for handle

numeric_cols = ['Year', 'Matches_Batted', 'Not_Outs', 'Runs_Scored', 'Highest_Score', 
                'Batting_Average', 'Balls_Faced', 'Batting_Strike_Rate', 'Centuries', 
                'Half_Centuries', 'Fours', 'Sixes', 'Catches_Taken', 'Stumpings', 
                'Matches_Bowled', 'Balls_Bowled', 'Runs_Conceded', 'Wickets_Taken', 
                'Bowling_Average', 'Economy_Rate', 'Bowling_Strike_Rate', 
                'Four_Wicket_Hauls', 'Five_Wicket_Hauls']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
df.info()
```

```python
df.isnull().sum()
```

Who scored the most runs in 2024?

```python
# Filter for 2024
ipl_2024 = df[df["Year"] == 2024]

# Find top 10 run scorers
top_10_scorers = ipl_2024.sort_values("Runs_Scored", ascending=False).head(10)

print("\nTop 5 run scorers in 2024:")
print(top_10_scorers[['Player_Name', 'Runs_Scored']])
```

Who took the most wickets in 2024

```python
# Find to 10 Wickets takers
top_10_bowlers = ipl_2024.sort_values("Wickets_Taken", ascending=False).head(10)

print("\nTop 5 run wicket taken in 2024:")
print(top_10_bowlers[['Player_Name', 'Wickets_Taken']])
```

## Simple visualization 

Top Run Scorers in 2024

```python
# create a simple chart

plt.figure(figsize=(12,8))
plt.barh(top_10_scorers['Player_Name'], top_10_scorers['Runs_Scored'], color='b')

font1 = {'family': 'serif', 'color':'blue', 'size':30}
font2 = {'family': 'serif', 'color':'green', 'size':20}

plt.title('Top 10 Run Scorers in IPL 2024', fontdict= font1)
plt.xlabel('Runs Scored', fontdict=font2)
plt.gca().invert_yaxis()  # Shows highest at top
plt.show()
```

Top Wicket Takers in 2024

```python
plt.figure(figsize=(12,8))
plt.barh(top_10_bowlers['Player_Name'], top_10_bowlers['Wickets_Taken'], color='green')

font1 = {'family': 'serif', 'color':'blue', 'size':30}
font2 = {'family': 'serif', 'color':'red', 'size':20}

plt.title('Top 10 wickets taken in IPL 2024', fontdict= font1)
plt.xlabel('Wickets Taken', fontdict=font2)
plt.gca().invert_yaxis()  # Shows highest at top
plt.show()
```

```python
# Get data for a specific player (example: Virat Kohli)
kohli_data = df[(df['Player_Name'] == 'Virat Kohli') & (df["Year"]>=2008) & (df["Year"]<=2024)].sort_values('Year')

plt.figure(figsize=(9, 6))
plt.plot(kohli_data['Year'], kohli_data['Runs_Scored'], marker='o',color='green', label='Runs Scored')
plt.plot(kohli_data['Year'], kohli_data['Batting_Average'], marker='s',color='red', label='Batting Average')
plt.xlabel('Year')
plt.ylabel('Performance Metric')
plt.title('Virat Kohli Performance Over Years')
plt.xticks(range(2008, 2025))
plt.grid()
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
