```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("imdb.csv")
df.head()
```

```python
## Remove 'min' from the Runtime column and convert it to integer.
df["Runtime"].head()
df["Runtime"] = df["Runtime"].str.replace(" min", "", regex=True).astype(int)
```

```python
df["Runtime"]
```

```python
df.info()
```

```python
df["Gross"]
```

```python
## Clean the Gross column by removing commas and converting it to numbers.
# Remove commas from the 'Gross' column and convert to float
df["Gross"].str.replace(",","").astype(float)
```

```python
## Convert the Released Year column to numeric data.
## Convert the 'Released Year' column to numeric, coercing errors to NaN
df["Released_Year"] = pd.to_numeric(df["Released_Year"], errors="coerce")
```

```python
df["Released_Year"]
```

```python
df["Certificate"].isnull().sum()
```

```python
## Replace missing values in the Certificate column with 'Not Rated'.
df["Certificate"].fillna(value='Not Rated', inplace=True)
df["Certificate"]
```

```python
df['Meta_score']
df['Meta_score'].isnull().sum()
## Replace missing values in the Meta_score column with the average metascore.
df['Meta_score'].fillna(value='Average_Metascore', inplace=True)
df[df['Meta_score'] == 'Average_Metascore'].head(5)
```

```python
df['Meta_score'].isnull().sum()
```

```python
## Create a new column 'Normalized_Rating' by normalizing the IMDB rating between 0 and 1 look out the formula for Normalization.
min_rating = df["IMDB_Rating"].min()
min_rating
max_rating = df["IMDB_Rating"].max()
max_rating
df["Normalized_Rating"] = (df["IMDB_Rating"]-min_rating)/(max_rating - min_rating)
# df.insert(loc= 7, value= df["Normalized_Rating"], column="Normalizing the IMDB rating between 0 and 1" )
print(df["Normalized_Rating"])
```
